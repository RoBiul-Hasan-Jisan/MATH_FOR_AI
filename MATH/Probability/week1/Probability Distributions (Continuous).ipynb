f(x) ≥ 0

Cumulative Distribution Function (CDF)
F(x) = P(X ≤ x) = ∫_{-∞}ˣ f(t)dt

Key Properties
Mean (μ): E[X] = ∫xf(x)dx

Variance (σ²): E[(X-μ)²] = ∫(x-μ)²f(x)dx

Skewness & Kurtosis: Shape characteristics

Essential Continuous Distributions
1. Normal (Gaussian) Distribution
Most important distribution in statistics/ML

PDF: f(x) = (1/√(2πσ²))e^{-(x-μ)²/(2σ²)}

Parameters: μ (mean), σ (standard deviation)

68-95-99.7 Rule: ±1σ (68%), ±2σ (95%), ±3σ (99.7%)

Applications:

Central Limit Theorem foundation

Error terms in linear regression

Many natural phenomena

Prior distributions in Bayesian methods

2. Standard Normal Distribution
Special case: μ=0, σ=1

Z-scores: (x-μ)/σ

Basis for hypothesis testing

3. Log-Normal Distribution
Y = e^X where X ~ Normal

Applications: Income distributions, stock prices, response times

4. Exponential Distribution
PDF: f(x) = λe^{-λx} for x ≥ 0

Memoryless property: P(X>s+t|X>s) = P(X>t)

Applications: Time between events, survival analysis

5. Gamma Distribution
Generalization of Exponential

Parameters: shape (k), scale (θ)

Applications: Wait times for multiple events, Bayesian priors

6. Beta Distribution
PDF: f(x) = [x^{α-1}(1-x)^{β-1}]/B(α,β) for x∈[0,1]

Parameters: α, β > 0

Applications:

Modeling probabilities/proportions

Bayesian inference (conjugate prior for binomial)

A/B testing

7. Uniform Distribution
PDF: f(x) = 1/(b-a) for x∈[a,b]

Applications: Random initialization, non-informative priors

8. Chi-Squared (χ²) Distribution
Sum of squared standard normals

Applications: Hypothesis testing, variance estimation

9. Student's t-Distribution
Heavier tails than normal

Applications: Small sample inference, robust statistics

10. F-Distribution
Ratio of two chi-squared variables

Applications: ANOVA, regression analysis

11. Pareto Distribution
Power law distribution

PDF: f(x) = αxₘᵃ/x^{α+1} for x ≥ xₘ

Applications: Wealth distribution, city sizes, network phenomena

12. Laplace (Double Exponential) Distribution
PDF: f(x) = (1/(2b))e^{-|x-μ|/b}

Applications: Robust regression, Bayesian inference, sparse signals

Multivariate Distributions
Multivariate Normal
PDF involves covariance matrix Σ

Key property: Marginal and conditional distributions are also normal

Applications: Gaussian processes, factor analysis

Dirichlet Distribution
Multivariate generalization of Beta

Applications: Topic modeling (Latent Dirichlet Allocation), categorical data priors

Applications in ML/AI
1. Bayesian Methods
Priors and posteriors in Bayesian inference

Conjugate priors:

Beta → Binomial

Dirichlet → Multinomial

Normal → Normal (with known variance)

Gamma → Poisson

2. Generative Models
GANs, VAEs use various distributions

Mixture models (Gaussian Mixture Models)

Normalizing flows for density estimation

3. Statistical Learning
Linear Regression: Errors typically normal

GLMs: Exponential family distributions

Survival Analysis: Exponential, Weibull distributions

4. Anomaly Detection
Model normal behavior with distributions

Flag low-probability events

5. Reinforcement Learning
Policy distributions

Exploration strategies

6. Natural Language Processing
Topic modeling (Dirichlet)

Word embeddings (sometimes modeled as Gaussian)

Key Theorems & Concepts
Central Limit Theorem
Sample means approach normal distribution

Foundation for many statistical methods

Law of Large Numbers
Sample means converge to population mean

Exponential Family
General form: p(x|θ) = h(x)exp[η(θ)·T(x)-A(θ)]

Includes: Normal, Exponential, Beta, Gamma, Poisson, etc.

Nice mathematical properties for ML

Practical Considerations
Distribution Selection Guidelines
Domain knowledge: What distribution makes sense?

Data exploration: Histograms, Q-Q plots

Statistical tests: Kolmogorov-Smirnov, Shapiro-Wilk

Model requirements: Conjugacy, computational tractability

Parameter Estimation Methods
Maximum Likelihood Estimation (MLE)

Method of Moments

Bayesian Estimation (with priors)

Transformations
Box-Cox: Handle non-normality

Log transform: Right-skewed data

Z-score standardization: For comparability
