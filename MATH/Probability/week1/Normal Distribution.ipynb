# Normal (Gaussian) distribution 
The Normal (Gaussian) distribution is the most important probability distribution in statistics, describing many natural phenomena and forming the foundation of statistical inference.
```bash
Mathematical Definition
For parameters μ (mean) and σ² (variance), where σ > 0:

PDF (Probability Density Function):


f(x) = (1/√(2πσ²)) * exp(-(x-μ)²/(2σ²))
CDF (Cumulative Distribution Function):


Φ(x) = (1/√(2πσ²)) ∫_{-∞}^{x} exp(-(t-μ)²/(2σ²)) dt
No closed form; computed numerically


Normal PDF (Bell Curve):                  Normal CDF (Sigmoid Shape):
      ▲                                       ▲
      │                                       │ 1.0
      │               _                       │       ┌──────
      │             /   \                     │      /
      │           /       \                   │     /
      │         /           \                 │    /
      │       /               \               │   /
      │     /                   \             │  /
      │   /                       \           │ /
      │ /                           \         │/
      └──────┴─────┴─────┴─────┴─────► x      └───────────────► x
          μ-2σ μ-σ  μ  μ+σ μ+2σ                  μ

Properties:
- Symmetric about μ
- Bell-shaped
- Tails extend to ±∞
- S-shaped CDF


Standard Normal Distribution
Special Case: μ=0, σ=1
Standard Normal (Z): N(0, 1)

PDF: φ(z) = (1/√(2π)) * e^{-z²/2}

CDF: Φ(z) = ∫_{-∞}^{z} φ(t) dt

Z-score Standardization
Any normal variable can be standardized:


Z = (X - μ) / σ ~ N(0, 1)

# Reverse transformation:
X = μ + σ * Z
68-95-99.7 (Empirical) Rule

P(μ - σ ≤ X ≤ μ + σ) ≈ 0.6827   # ≈68%
P(μ - 2σ ≤ X ≤ μ + 2σ) ≈ 0.9545  # ≈95%
P(μ - 3σ ≤ X ≤ μ + 3σ) ≈ 0.9973  # ≈99.7%



Moments

Mean = μ
Median = μ
Mode = μ

Variance = σ²
Standard Deviation = σ

Skewness = 0          # Perfectly symmetric
Kurtosis = 3          # Excess Kurtosis = 0

# All odd central moments = 0 (symmetry)
# Even central moments: μ_{2k} = (2k)!/(2^k k!) σ^{2k}


Characteristic Function

φ(t) = exp(iμt - ½σ²t²)
Moment Generating Function

M(t) = exp(μt + ½σ²t²)
```
