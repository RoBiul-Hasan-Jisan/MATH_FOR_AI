# Probability Density Function
A Probability Density Function (PDF) is a function f(x) that describes the relative likelihood of a continuous random variable taking on a given value.
```bash
Mathematical Definition
For a continuous random variable X, the PDF f(x) satisfies:

Non-negativity: f(x) ≥ 0 for all x

Normalization: ∫_{-∞}^{∞} f(x) dx = 1

Probability calculation: P(a ≤ X ≤ b) = ∫_{a}^{b} f(x) dx

Key Insight
Unlike discrete probabilities, f(x) is NOT probability for continuous variables:

f(x) can be > 1 (density, not probability)

P(X = x) = 0 for any specific x (continuous case)

We only talk about probabilities over intervals

Density Visualization:
      ▲
      │
      │         ∧
      │        / \
      │       /   \
f(x)  │      /     \
      │     /       \
      │    /         \
      │   /           \
      └──┴─────┴─────┴───► x
         a     c     b
       
P(a ≤ X ≤ b) = Area under curve between a and b


Area Interpretation
Total area under PDF curve = 1 (100% probability)

Area between a and b = Probability X falls in [a, b]

No area at a point (line has zero width)


Standard Normal Distribution

f(x) = (1/√(2π)) * e^(-x²/2)

# Properties:
# - Bell-shaped curve
# - Symmetric about 0
# - Mean = Median = Mode = 0


Exponential Distribution (λ > 0)

f(x) = λ * e^(-λx) for x ≥ 0, 0 otherwise

# Properties:
# - Decreasing function
# - Memoryless: P(X>s+t | X>s) = P(X>t)

Uniform Distribution (a < b)

f(x) = 1/(b-a) for a ≤ x ≤ b, 0 otherwise

# Properties:
# - Constant height
# - All intervals of same width have equal probability


Relationship with CDF
F(x) = ∫_{-∞}^{x} f(t) dt    # CDF from PDF
f(x) = dF(x)/dx             # PDF from CDF (if F differentiable)



Expectation using PDF

E[X] = ∫_{-∞}^{∞} x * f(x) dx
E[g(X)] = ∫_{-∞}^{∞} g(x) * f(x) dx

# Variance:
Var(X) = E[X²] - (E[X])²
       = ∫ (x-μ)² f(x) dx


Change of Variables
If Y = g(X) and g is monotonic:


f_Y(y) = f_X(g⁻¹(y)) * |d/dy g⁻¹(y)|
```
