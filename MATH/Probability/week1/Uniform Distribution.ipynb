A continuous uniform distribution describes an experiment where all outcomes in a given interval are equally likely.
```bash
Mathematical Definition
For parameters a (minimum) and b (maximum), where a < b:

PDF (Probability Density Function):


f(x) = {
    1/(b-a)  for a ≤ x ≤ b,
    0        otherwise
}
CDF (Cumulative Distribution Function):


F(x) = {
    0                    for x < a,
    (x-a)/(b-a)         for a ≤ x ≤ b,
    1                    for x > b
}



Uniform PDF:                    Uniform CDF:
      ▲                            ▲
      │                            │ 1.0
      │        _________           │      /
      │       |         |          │     /
      │       |         |          │    /
f(x)  │       |         |          │   /
      │       |         |          │  /
      │       |_________|          │ /
      │                            │/
      └──────┴─────┴─────┴──► x    └──────┴─────┴─────┴──► x
            a     μ     b                a          b

Properties:
- Constant height
- Total area = 1
- Linear CDF


Parameters
a: Minimum value (lower bound)

b: Maximum value (upper bound)

Key Statistics

Mean (μ) = (a + b) / 2
Median = (a + b) / 2  # Same as mean
Mode = Any value in [a, b]  # Not unique

Variance (σ²) = (b - a)² / 12
Standard Deviation (σ) = (b - a) / √12 ≈ (b - a) / 3.464

Skewness = 0  # Symmetric
Kurtosis = -6/5 = -1.2  # Platykurtic (flatter than normal)
Special Case: Standard Uniform
When a = 0, b = 1:


f(x) = 1 for 0 ≤ x ≤ 1
μ = 0.5
σ² = 1/12 ≈ 0.0833
3. Probability Calculations
Interval Probabilities

# For any interval [c, d] within [a, b]:
P(c ≤ X ≤ d) = (d - c) / (b - a)

# Examples:
P(a ≤ X ≤ b) = 1
P(X ≤ m) = (m - a)/(b - a)  where m = (a+b)/2 is median
P(X > x) = (b - x)/(b - a)
Conditional Probabilities

# Given X > c, probability X > d:
P(X > d | X > c) = (b - d)/(b - c)  for c ≤ d ≤ b

# Memoryless property? NO!
# Uniform is NOT memoryless (unlike Exponential)

```
