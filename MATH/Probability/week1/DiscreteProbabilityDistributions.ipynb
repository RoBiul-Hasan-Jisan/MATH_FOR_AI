# discrete probability distribution 
A discrete probability distribution describes the probability of occurrence of each value of a discrete random variable. It consists of all possible values and their associated probabilities.

## Formal Definition
For a discrete random variable X, the probability distribution is defined by:
```bash
Set of possible values: x₁, x₂, ..., xₙ

Probability mass function: p(x) = P(X = x)

Conditions:

0 ≤ p(x) ≤ 1 for all x

∑ p(x) = 1 (sum over all possible x)

Common Discrete Distributions
1. Discrete Uniform Distribution
Description: All outcomes are equally likely

PMF: p(x) = 1/n for x = 1, 2, ..., n

Parameters: n (number of outcomes)

Example: Fair die roll

p(1) = p(2) = ... = p(6) = 1/6

Mean: E[X] = (n + 1)/2
Variance: Var(X) = (n² - 1)/12

Die example: E[X] = (6 + 1)/2 = 3.5, Var(X) = (36 - 1)/12 ≈ 2.92

2. Bernoulli Distribution
Description: Single trial with two outcomes (success/failure)

PMF:

p(1) = p (success)

p(0) = 1 - p (failure)

Parameters: p (probability of success)

Example: Coin flip (p = 0.5), Quality control (defective/not)

Mean: E[X] = p
Variance: Var(X) = p(1 - p)

Example: p = 0.3 → E[X] = 0.3, Var(X) = 0.3 × 0.7 = 0.21

3. Binomial Distribution
Description: Number of successes in n independent Bernoulli trials

PMF: P(X = k) = C(n,k) × p^k × (1-p)^(n-k)
where C(n,k) = n!/(k!(n-k)!)

Parameters: n (number of trials), p (success probability)

Notation: X ~ Binomial(n, p)

Example: Number of heads in 10 coin flips

Mean: E[X] = np
Variance: Var(X) = np(1-p)

Detailed Example:
Problem: Fair coin flipped 5 times. Find probability of exactly 3 heads.

Solution:

n = 5, p = 0.5, k = 3

P(X = 3) = C(5,3) × (0.5)³ × (0.5)²

C(5,3) = 10

P(X = 3) = 10 × (1/8) × (1/4) = 10/32 = 0.3125

Complete Distribution (n=5, p=0.5):

k	P(X = k)	Calculation
0	0.03125	1 × (0.5)⁰ × (0.5)⁵
1	0.15625	5 × (0.5)¹ × (0.5)⁴
2	0.31250	10 × (0.5)² × (0.5)³
3	0.31250	10 × (0.5)³ × (0.5)²
4	0.15625	5 × (0.5)⁴ × (0.5)¹
5	0.03125	1 × (0.5)⁵ × (0.5)⁰
Verification: Sum = 1.00000 

4. Poisson Distribution
Description: Number of events occurring in a fixed interval of time/space

PMF: P(X = k) = (λ^k × e^(-λ)) / k!

Parameters: λ (average rate of events)

Notation: X ~ Poisson(λ)

Example: Number of customers arriving per hour, Number of defects per unit

Mean: E[X] = λ
Variance: Var(X) = λ

Detailed Example:
Problem: Calls arrive at a call center at average rate of 4 per hour. Find probability of exactly 6 calls in next hour.

Solution:

λ = 4, k = 6

P(X = 6) = (4⁶ × e^(-4)) / 6!

4⁶ = 4096, 6! = 720, e^(-4) ≈ 0.0183

P(X = 6) = (4096 × 0.0183) / 720 ≈ 0.1042

Complete Distribution (λ=4):

k	P(X = k)
0	0.0183
1	0.0733
2	0.1465
3	0.1954
4	0.1954
5	0.1563
6	0.1042
7	0.0595
5. Geometric Distribution
Description: Number of trials until first success

PMF: P(X = k) = (1-p)^(k-1) × p for k = 1, 2, 3, ...

Parameters: p (success probability)

Example: Number of coin flips until first head

Mean: E[X] = 1/p
Variance: Var(X) = (1-p)/p²

Detailed Example:
Problem: Die rolled until first 6 appears. Find probability it takes exactly 4 rolls.

Solution:

p = 1/6, k = 4

P(X = 4) = (5/6)³ × (1/6) = (125/216) × (1/6) = 125/1296 ≈ 0.0965

6. Negative Binomial Distribution
Description: Number of trials until r-th success

PMF: P(X = k) = C(k-1, r-1) × p^r × (1-p)^(k-r) for k = r, r+1, ...

Parameters: r (number of successes), p (success probability)

Example: Number of coin flips until 3rd head

Mean: E[X] = r/p
Variance: Var(X) = r(1-p)/p²

7. Hypergeometric Distribution
Description: Number of successes in n draws without replacement from finite population

PMF: P(X = k) = [C(K,k) × C(N-K, n-k)] / C(N,n)

Parameters: N (population size), K (successes in population), n (sample size)

Example: Drawing cards from deck, quality control sampling

Detailed Example:
Problem: Deck of 52 cards, draw 5 cards. Find probability of exactly 2 Aces.

Solution:

N = 52, K = 4 (Aces), n = 5, k = 2

P(X = 2) = [C(4,2) × C(48,3)] / C(52,5)

C(4,2) = 6, C(48,3) = 17296, C(52,5) = 2598960

P(X = 2) = (6 × 17296) / 2598960 ≈ 0.0399

Distribution Comparison Table
Distribution	PMF	Parameters	When to Use
Uniform	1/n	n	All outcomes equally likely
Bernoulli	p(x) = p if x=1, 1-p if x=0	p	Single yes/no trial
Binomial	C(n,k)p^k(1-p)^(n-k)	n, p	Fixed trials, constant p, independent
Poisson	(λ^k e^(-λ))/k!	λ	Rare events, counts in interval
Geometric	(1-p)^(k-1)p	p	Trials until first success
Negative Binomial	C(k-1,r-1)p^r(1-p)^(k-r)	r, p	Trials until r-th success
Hypergeometric	[C(K,k)C(N-K,n-k)]/C(N,n)	N, K, n	Sampling without replacement
Practice Problems
A fair coin is flipped 8 times. What's the probability of at least 6 heads?

X ~ Binomial(8, 0.5)

P(X ≥ 6) = P(X=6) + P(X=7) + P(X=8)

= C(8,6)(0.5)⁸ + C(8,7)(0.5)⁸ + C(8,8)(0.5)⁸

= (28 + 8 + 1) × (1/256) = 37/256 ≈ 0.1445

Cars arrive at toll booth at rate 10 per minute. What's probability exactly 12 cars in next minute?

X ~ Poisson(10)

P(X = 12) = (10¹² × e^(-10)) / 12! ≈ 0.0948

Draw 3 cards from 52-card deck without replacement. Find expected number of Aces.

X ~ Hypergeometric(N=52, K=4, n=3)

E[X] = n × (K/N) = 3 × (4/52) = 12/52 ≈ 0.2308
```
