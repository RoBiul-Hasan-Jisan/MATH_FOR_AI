The Cumulative Distribution Function F(x) gives the probability that a random variable X takes a value less than or equal to x:

F(x) = P(X ≤ x)

Mathematical Properties (for any random variable)
Monotonicity: F(x) is non-decreasing: if x₁ < x₂ then F(x₁) ≤ F(x₂)
```bash
Limits:

lim_{x→-∞} F(x) = 0

lim_{x→∞} F(x) = 1

Right-continuity: lim_{h→0⁺} F(x+h) = F(x) (for all x)

For Continuous Random Variables
If X is continuous with PDF f(x):


F(x) = ∫_{-∞}^{x} f(t) dt          # CDF from PDF
f(x) = dF(x)/dx                  # PDF from CDF (where derivative exists)
For Discrete Random Variables
If X is discrete with PMF p(x):


F(x) = Σ_{t ≤ x} p(t)       # Sum over all values ≤ x




CDF Visualization for Continuous Variable:

      ▲
      │ 1.0
      │     ┌─────────
      │    /
      │   /
F(x)  │  /
      │ /
      │/
      └────────────────► x
     
Properties:
- Starts at 0 on left, ends at 1 on right
- Never decreases
- Smooth for continuous variables
- Jumps for discrete variables


Comparison: Continuous vs Discrete CDF

```bash
      ▲                    ▲
      │ 1.0                │ 1.0
      │     ┌───────       │     │
      │    /               │     │
      │   /                │     │
F(x)  │  /                 │     │
      │ /                  │   │ │
      │/                   │   │ │
      └───────► x          └───┴─┴──► x
      Continuous           Discrete
      (Smooth)             (Step function)



Probability Calculations using CDF

# Basic probability
P(X ≤ b) = F(b)

# Interval probabilities
P(a < X ≤ b) = F(b) - F(a)          # Most common
P(a ≤ X ≤ b) = F(b) - F(a) + P(X=a) # For discrete case

# Complement rule
P(X > x) = 1 - F(x)

# Point probabilities (discrete only)
P(X = x) = F(x) - F(x⁻)             # Jump at x
# For continuous: P(X = x) = 0
Quantiles and Percentiles

# p-th quantile: Q(p) = F⁻¹(p) = min{x: F(x) ≥ p}
# Special cases:
Median = Q(0.5) = F⁻¹(0.5)
First Quartile Q1 = Q(0.25)
Third Quartile Q3 = Q(0.75)

# Interquartile Range (IQR) = Q3 - Q1
Survival Function

S(x) = 1 - F(x) = P(X > x)
# Used in reliability engineering, survival analysis


Standard Normal Distribution

F(x) = Φ(x) = (1/√(2π)) ∫_{-∞}^{x} e^{-t²/2} dt

# Properties:
# - No closed form, use tables/software
# - Φ(0) = 0.5 (symmetric)
# - Φ(-x) = 1 - Φ(x)
Exponential Distribution (λ > 0)

F(x) = 1 - e^{-λx} for x ≥ 0, 0 otherwise

# Closed form available
# Shows memoryless property: P(X>s+t | X>s) = P(X>t)
Uniform Distribution (a < b)

F(x) = 0                      for x < a
     = (x-a)/(b-a)            for a ≤ x ≤ b
     = 1                      for x > b

# Piecewise linear function
```
