For any probability experiment, if you add up the probabilities of all possible outcomes in the sample space, the sum must equal 1.
```bash
Formula: ∑ P(All Outcomes) = 1

Examples:
Coin: P(Heads) + P(Tails) = 0.5 + 0.5 = 1

Die: P(1) + P(2) + P(3) + P(4) + P(5) + P(6) = 6 × (1/6) = 1

Deck of Cards: Sum of probabilities for all 52 cards = 1

Key Rule #1: The Addition Rule for Mutually Exclusive Events
Mutually Exclusive Events = Events that cannot happen at the same time.

Formula: P(A ∪ B) = P(A) + P(B)
(Where ∪ means "or")

Examples:
With a Die:

Event A: Roll a 2 → P(A) = 1/6

Event B: Roll a 5 → P(B) = 1/6

P(A or B) = P(A) + P(B) = 1/6 + 1/6 = 2/6 = 1/3

With Cards:

Event A: Draw a Heart → P(A) = 13/52

Event B: Draw a Spade → P(B) = 13/52

P(Heart or Spade) = 13/52 + 13/52 = 26/52 = 1/2

Key Rule #2: The General Addition Rule (For Any Events)
For any two events (whether mutually exclusive or not):

Formula: P(A ∪ B) = P(A) + P(B) - P(A ∩ B)

Why subtract P(A ∩ B)? Because when we add P(A) + P(B), we double-count the outcomes where both A and B occur.

Examples:
With a Die:

Event A: Roll an even number {2, 4, 6} → P(A) = 3/6

Event B: Roll a number greater than 3 {4, 5, 6} → P(B) = 3/6

A ∩ B: {4, 6} → P(A ∩ B) = 2/6

P(A or B) = 3/6 + 3/6 - 2/6 = 4/6 = 2/3

With Cards:

Event A: Draw a Face Card (12 cards) → P(A) = 12/52

Event B: Draw a Heart (13 cards) → P(B) = 13/52

A ∩ B: Face Card that is a Heart (3 cards) → P(A ∩ B) = 3/52

P(A or B) = 12/52 + 13/52 - 3/52 = 22/52 = 11/26

Key Rule #3: Probability of Complementary Events
The complement of event A (written as A') is "everything that is NOT A".

Formula: P(A) + P(A') = 1
or P(A') = 1 - P(A)

This is extremely useful for solving "at least one" problems.

Examples:
With Coins:

P(At least one Head in 3 flips) = 1 - P(No Heads) = 1 - P(TTT) = 1 - (1/8) = 7/8

With Dice:

P(Roll at least 3) = 1 - P(Roll less than 3) = 1 - P(1 or 2) = 1 - 2/6 = 4/6 = 2/3

With Cards:

P(Not an Ace) = 1 - P(Ace) = 1 - 4/52 = 48/52 = 12/13

Special Cases and Important Notes
For Mutually Exclusive Events:
If A and B are mutually exclusive, then P(A ∩ B) = 0, so:
P(A or B) = P(A) + P(B)

For Three Events:
P(A or B or C) = P(A) + P(B) + P(C) - P(A∩B) - P(A∩C) - P(B∩C) + P(A∩B∩C)

Probability Range:
Any probability must be between 0 and 1:
0 ≤ P(A) ≤ 1

Practice Problems
What is P(Rolling a 1 or 6 on a die)?

Mutually exclusive: 1/6 + 1/6 = 2/6 = 1/3

What is P(Drawing a Heart or a King from a deck)?

P(Heart) = 13/52, P(King) = 4/52, P(King of Hearts) = 1/52

P(Heart or King) = 13/52 + 4/52 - 1/52 = 16/52 = 4/13

If P(Rain) = 0.3, what is P(No Rain)?

P(No Rain) = 1 - 0.3 = 0.7

In a class, P(Pass Math) = 0.8, P(Pass English) = 0.9, P(Pass Both) = 0.75. What is P(Pass at least one)?

P(Math or English) = 0.8 + 0.9 - 0.75 = 0.95
```
