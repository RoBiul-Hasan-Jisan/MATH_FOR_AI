# Conditional probability
Conditional probability is the probability of an event occurring given that another event has already occurred. It's a fundamental concept for understanding dependent events.

## Formal Definition
The conditional probability of A given B is:
P(A|B) = P(A ∩ B) / P(B), provided P(B) > 0

Where:

P(A|B) = "Probability of A given B"

P(A ∩ B) = Probability that both A and B occur

P(B) = Probability of event B

Key Interpretation
P(A|B) answers the question: "If I know that B has happened, what is the probability that A also happens?"

Visual Understanding
Think of reducing our sample space to only outcomes where B occurs

We're looking at what fraction of B's outcomes also satisfy A

Examples with Coins, Dice, and Cards
Coin Examples
Example 1: Two coin flips

B: At least one Head occurs

A: Both flips are Heads

What is P(A|B)?

Solution:

Sample space: {HH, HT, TH, TT}

P(B) = P(at least one Head) = 3/4

P(A ∩ B) = P(both Heads) = 1/4

P(A|B) = (1/4) / (3/4) = 1/3

Dice Examples
Example 2: Single die roll

B: Number rolled is even {2, 4, 6}

A: Number is greater than 3 {4, 5, 6}

What is P(A|B)?

Solution:

P(B) = 3/6 = 1/2

A ∩ B = {4, 6} → P(A ∩ B) = 2/6 = 1/3

P(A|B) = (1/3) / (1/2) = 2/3

Example 3: Two dice

B: Sum of two dice is 8

A: First die shows 3

What is P(A|B)?

Solution:

Ways to get sum of 8: (2,6), (3,5), (4,4), (5,3), (6,2) → 5 outcomes

P(B) = 5/36

A ∩ B: First die=3 AND sum=8 → (3,5) → 1 outcome

P(A ∩ B) = 1/36

P(A|B) = (1/36) / (5/36) = 1/5

Card Examples
Example 4: Drawing cards without replacement

Draw 2 cards from a deck

B: First card is an Ace

A: Second card is an Ace

What is P(A|B)?

Solution:

If first card is Ace: 51 cards remain, 3 Aces left

P(A|B) = 3/51

Example 5: Conditional on suit

B: Card is a Face Card (Jack, Queen, King)

A: Card is a Heart

What is P(A|B)?

Solution:

Total face cards = 12 (4 suits × 3 face cards)

Face cards that are Hearts = 3

P(A|B) = 3/12 = 1/4

Important Relationships
1. Connection to Joint Probability
From the definition: P(A ∩ B) = P(B) × P(A|B)
This is incredibly useful for calculating joint probabilities.

2. Connection to Independence
If A and B are independent:

P(A|B) = P(A)

P(B|A) = P(B)

3. Bayes' Theorem
P(A|B) = [P(B|A) × P(A)] / P(B)
This allows us to "reverse" conditional probabilities.

Special Cases and Formulas
Law of Total Probability
If B₁, B₂, ..., Bₙ partition the sample space, then:
P(A) = P(A|B₁)P(B₁) + P(A|B₂)P(B₂) + ... + P(A|Bₙ)P(Bₙ)

Example of Law of Total Probability
There are 2 boxes:

Box 1: 3 red balls, 2 blue balls

Box 2: 1 red ball, 4 blue balls

Pick a box at random (50% each), then draw a ball

P(Red) = P(Red|Box1)P(Box1) + P(Red|Box2)P(Box2)
= (3/5)(1/2) + (1/5)(1/2) = 3/10 + 1/10 = 4/10 = 2/5

Common Mistakes to Avoid
Confusing P(A|B) with P(B|A)

P(Cancer|Positive Test) vs P(Positive Test|Cancer) are very different!

Assuming independence when it doesn't exist

Cards without replacement are dependent

Using conditional probability when P(B) = 0

P(A|B) is undefined if P(B) = 0

Practice Problems
In a family with two children, if at least one is a boy, what's P(Both are boys)?

Sample space: {BB, BG, GB, GG}

B: At least one boy = {BB, BG, GB} → P(B) = 3/4

A: Both boys = {BB} → P(A ∩ B) = 1/4

P(A|B) = (1/4) / (3/4) = 1/3

Roll a fair die. Given that the number is prime, what's P(Number > 3)?

Prime numbers: {2, 3, 5} → P(prime) = 3/6

Prime numbers > 3: {5} → P(prime and >3) = 1/6

P(>3 | prime) = (1/6) / (3/6) = 1/3

From a deck, if a card is red, what's P(It is a Heart)?

P(Red) = 26/52 = 1/2

P(Red and Heart) = P(Heart) = 13/52 = 1/4

P(Heart | Red) = (1/4) / (1/2) = 1/2
