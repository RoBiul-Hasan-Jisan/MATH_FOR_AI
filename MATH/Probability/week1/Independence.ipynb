# Independence
Independence is a crucial concept where the occurrence of one event does not affect the probability of another event occurring.

## Formal Definition
Two events A and B are independent if and only if:
P(A ∩ B) = P(A) × P(B)

This also implies:

P(A|B) = P(A)

P(B|A) = P(B)
```bash
Key Characteristics of Independent Events
1. No Influence
The outcome of one event has no impact on the other

Knowing that B occurred gives no information about whether A occurred

2. Conditional Probability Equals Marginal Probability
P(A|B) = P(A)

P(B|A) = P(B)

3. Symmetry
If A is independent of B, then B is independent of A

Examples of Independent Events
Coin Examples
Example 1: Multiple coin flips

Flip a fair coin 3 times

P(Heads on 1st) = 1/2

P(Heads on 2nd) = 1/2

P(Heads on 3rd) = 1/2

These are independent: P(H1 ∩ H2 ∩ H3) = (1/2) × (1/2) × (1/2) = 1/8

Example 2: Different coins

Flip two different coins

P(Coin A shows Heads AND Coin B shows Tails) = (1/2) × (1/2) = 1/4

Dice Examples
Example 3: Rolling two dice

Roll a red die and a blue die

P(Red die shows 3) = 1/6

P(Blue die shows 5) = 1/6

P(Red=3 AND Blue=5) = (1/6) × (1/6) = 1/36

Example 4: Multiple rolls of same die

Roll a fair die twice

P(First roll is 2) = 1/6

P(Second roll is 6) = 1/6

P(First=2 AND Second=6) = (1/6) × (1/6) = 1/36

Card Examples
Example 5: Drawing with replacement

Draw a card, note it, put it back, draw again

P(First card is Ace) = 4/52

P(Second card is King) = 4/52

P(First=Ace AND Second=King) = (4/52) × (4/52) = 16/2704

Examples of Dependent Events
Card Examples
Example 6: Drawing without replacement

Draw a card, keep it out, draw another

P(First card is Ace) = 4/52

P(Second card is Ace | First was Ace) = 3/51

Since 3/51 ≠ 4/52, these events are dependent

Example 7: Related characteristics

Event A: Card is a King → P(A) = 4/52

Event B: Card is a Heart → P(B) = 13/52

P(A ∩ B) = P(King of Hearts) = 1/52

Check: P(A) × P(B) = (4/52) × (13/52) = 52/2704 = 1/52

These are independent! (Surprising but true)

Testing for Independence
To check if events A and B are independent, verify any of these:

P(A ∩ B) = P(A) × P(B)

P(A|B) = P(A)

P(B|A) = P(B)

Practice Testing:
Test 1: Dice example

A: Roll even number on die {2,4,6} → P(A) = 3/6

B: Roll number > 3 {4,5,6} → P(B) = 3/6

A ∩ B: {4,6} → P(A ∩ B) = 2/6

P(A) × P(B) = (3/6) × (3/6) = 9/36 = 1/4

2/6 = 1/3 ≠ 1/4 → Dependent

Test 2: Card example

A: Card is red → P(A) = 26/52

B: Card is a Heart → P(B) = 13/52

A ∩ B: Card is red Heart → P(A ∩ B) = 13/52

P(A) × P(B) = (26/52) × (13/52) = 338/2704 = 0.125

13/52 = 0.25 ≠ 0.125 → Dependent

Special Cases and Common Mistakes
Mutually Exclusive vs. Independent
Mutually Exclusive: P(A ∩ B) = 0

Independent: P(A ∩ B) = P(A) × P(B)

If A and B are both mutually exclusive AND have positive probability, they cannot be independent

Multiple Independent Events
For events A, B, C to be independent, we need:

P(A ∩ B) = P(A) × P(B)

P(A ∩ C) = P(A) × P(C)

P(B ∩ C) = P(B) × P(C)

P(A ∩ B ∩ C) = P(A) × P(B) × P(C)

Real-World vs. Mathematical Independence
Mathematically: Defined by the formulas above

In real world: Use domain knowledge or data to determine

Practice Problems
Flip two coins. Are these independent?

P(Heads on first) = 1/2, P(Heads on second) = 1/2

P(Both Heads) = 1/4 = (1/2) × (1/2) → Independent

Draw two cards without replacement. Are these independent?

P(First is Ace) = 4/52

P(Second is Ace | First was Ace) = 3/51 ≠ 4/52 → Dependent

Roll a die. A: Get even number, B: Get number > 2. Independent?

P(A) = 3/6, P(B) = 4/6, P(A∩B) = 2/6

Check: (3/6)×(4/6) = 12/36 = 1/3 = 2/6 → Independent!

```
