```python
#lec 4. Systems of Linear Equations

#Representing Systems using Matrices
'''
x +  y +  z = 6  
2x + 3y + 7z = 20  
x + 3y + 4z = 14
'''
# Matrix Form (Augmented Matrix)
'''
[ 1  1  1 |  6 ]
[ 2  3  7 | 20 ]
[ 1  3  4 | 14 ]



Row Echelon Form (REF)
A matrix is in REF if:

All non-zero rows are above rows of all zeros.

Each leading entry of a row is to the right of the leading entry of the row above it.

All entries below a leading entry are zero.

[ 1  1  1 |  6 ]
[ 0  1  5 |  8 ]
[ 0  0  1 |  2 ]

Reduced Row Echelon Form (RREF)
A matrix is in RREF if it is in REF and:

The leading entry in each non-zero row is 1.

Each leading 1 is the only non-zero entry in its column.


[ 1  0  0 | 1 ]
[ 0  1  0 | 2 ]
[ 0  0  1 | 3 ]


'''

import sympy as sp

# Define the augmented matrix for the system
A = sp.Matrix([
    [1, 1, 1, 6],
    [2, 3, 7, 20],
    [1, 3, 4, 14]
])

# Compute the Reduced Row Echelon Form (RREF)
rref_matrix, pivot_columns = A.rref()

# Print the results
print("Reduced Row Echelon Form (RREF):")
sp.pprint(rref_matrix)

print("\nPivot Columns:", pivot_columns)



```

```python
import numpy as  np
'''
Gaussian Elimination
Goal: Solve a system of linear equations by converting the matrix to row echelon form (REF).

Steps:
Represent the system as an augmented matrix.

Use row operations (swap, scale, add rows) to make lower-left entries 0.

Once in row echelon form, use back substitution to find the solution.

2. Gauss-Jordan Elimination
Goal: Solve a system by converting the matrix to reduced row echelon form (RREF).

Difference from Gaussian:
Gaussian stops at upper triangular form, then back-substitutes.

Gauss-Jordan continues until matrix looks like the identity matrix (all pivots are 1 and other entries in pivot columns are 0).

3. Homogeneous vs Non-Homogeneous Systems
Homogeneous System:
All constants on the right-hand side are 0.

Form: Ax=0

Always has at least one solution: the trivial solution x=0

May also have infinitely many solutions (if system has free variables).

Non-Homogeneous System:
Constants on the right are not all 0.

Form:  Ax=b

May have no solution, one, or infinitely many.



'''

# Coefficient matrix A and RHS vector b

A = np.array([[1, 1, 1],
              [0, 2, 5],
              [2, 5, -1]], dtype=float)
b = np.array([6, -4, 27], dtype=float)
#Gaussian Elimination
# Augmented matrix
aug = np.hstack((A, b.reshape(-1, 1)))

# Gaussian Elimination to Row Echelon Form (REF)
n = len(b)
for i in range(n):
    # Pivoting
    max_row = np.argmax(abs(aug[i:, i])) + i
    aug[[i, max_row]] = aug[[max_row, i]]

    for j in range(i+1, n):
        factor = aug[j][i] / aug[i][i]
        aug[j] = aug[j] - factor * aug[i]

print("Row Echelon Form:\n", aug)

# Back substitution
x = np.zeros(n)
for i in range(n-1, -1, -1):
    x[i] = (aug[i][-1] - np.dot(aug[i][i+1:n], x[i+1:n])) / aug[i][i]

print("Solution from Gaussian Elimination:", x)




```

```python
#Lec 5: Matrix Properties

''' 
1. Rank of a Matrix
Definition: Number of linearly independent rows or columns.

Represents: The dimension of the column space (or row space).

Use: Determines if a system of linear equations has a unique solution.

2. Determinant
Definition: A scalar value that indicates if a matrix is invertible.

Use:

det ≠ 0: matrix is invertible.

det = 0: matrix is singular (non-invertible).

3. Trace
Definition: Sum of the diagonal elements of a square matrix.

Use: Appears in optimization, eigenvalue theory, and quantum mechanics.

 4. Null Space (Kernel)
Definition: Set of all vectors x such that A·x = 0.

Use: Represents the solution space to homogeneous equations.
5. Column Space & Row Space
Column Space: Set of all linear combinations of columns.

Row Space: Set of all linear combinations of rows.

Both are vector subspaces.

Use:

Help understand solutions to linear systems.

Their dimensions equal the rank.

6. Orthogonality and Orthonormality
Orthogonal: Two vectors are orthogonal if dot product = 0 (90° angle).

Orthonormal:

Vectors are orthogonal.

Each vector has unit length.
7. Linear Independence and Dependence
Independent: No vector can be written as a combination of others.

Dependent: At least one vector is a linear combination of others.

8. Span and Basis
Span: All vectors you can form using linear combinations of a set of vectors.

Basis:

A set of linearly independent vectors that span a space.

Basis vectors uniquely represent every vector in that space.

Example:

For 
𝑅^2 standard basis = [1,0], [0,1]


'''

import numpy as np

A = np.array([[1, 2], [2, 4]])
rank = np.linalg.matrix_rank(A)
print("Rank:", rank)

det = np.linalg.det(np.array([[1, 2], [3, 4]]))
print("Determinant:", det)


A = np.array([[1, 2], [3, 4]])
trace = np.trace(A)
print("Trace:", trace)  # Output: 1 + 4 = 5

from scipy.linalg import null_space

A = np.array([[2, 3], [4, 6]])
ns = null_space(A)
print("Null space:\n", ns)

u = np.array([3, 1])
v = np.array([1, 3])

print("Dot product:", np.dot(u, v))  # Output: 6 → not orthogonal


u = np.array([1, 0])
v = np.array([0, 1])

print("Dot product:", np.dot(u, v))  # Output: 0 → orthogonal


A = np.array([[1, 2], [2, 4]])
print("Rank:", np.linalg.matrix_rank(A))  # < number of columns ⇒ Dependent




```

```python
#lec 7   Eigenvalues and Eigenvectors

'''
1. Definition and Intuition
Linear Algebra studies vectors, matrices, and linear transformations.

It helps analyze systems of linear equations, transformations, and data structures.

Intuition: Think of matrices as functions that transform vectors in space — stretching, rotating, or projecting them.

2. Characteristic Polynomial
Given a square matrix A, the characteristic polynomial is defined as:

p(λ)=det(A−λI)
It’s a polynomial in λ whose roots are the eigenvalues of A.

Intuition: The values 

λ where A −𝜆𝐼 loses invertibility (determinant zero) correspond to directions that are only scaled (not rotated) by the transformation.

3. Diagonalization
A matrix A is diagonalizable if it can be written as:
A=PDP^−1
 
where D is a diagonal matrix of eigenvalues, and P contains eigenvectors as columns.
Intuition: Diagonalization simplifies transformations by converting to a basis where the matrix acts like simple scaling.

4. Geometric and Algebraic Multiplicity
Algebraic multiplicity: Number of times an eigenvalue appears as a root of the characteristic polynomial.

Geometric multiplicity: Dimension of the eigenspace corresponding to that eigenvalue (number of linearly independent eigenvectors).

Important: Geometric multiplicity ≤ Algebraic multiplicity.

Intuition: Algebraic multiplicity counts eigenvalue repetition; geometric multiplicity counts how many directions correspond to that eigenvalue.

5. Spectral Theorem
For a real symmetric matrix 
A, the spectral theorem states:
A can be diagonalized by an orthogonal matrix.

All eigenvalues of A are real.
There exists an orthonormal basis of eigenvectors.
Intuition: Symmetric matrices behave nicely — they can be "rotated" into diagonal form using orthogonal transformations.

Applications
6. Principal Component Analysis (PCA)
PCA uses eigenvalues/eigenvectors of the covariance matrix to find directions (principal components) that maximize variance.

Diagonalization of the covariance matrix extracts these principal components.

Intuition: Reduce data dimensionality by projecting onto most "important" axes.

7. Stability Analysis
Eigenvalues of the system matrix determine stability of dynamical systems.

If all eigenvalues have negative real parts → system is stable.

Intuition: Eigenvalues indicate growth/decay rates of system modes.

8. Markov Chains
Transition matrices have eigenvalue 1 with corresponding steady-state eigenvector.

Diagonalization helps analyze long-term behavior and convergence.

Intuition: Powering the matrix repeatedly projects any initial state to the steady state.
'''

import numpy as np
from scipy.integrate import solve_ivp
import matplotlib.pyplot as plt

A = np.array([[-2, 1],
              [0, -1]])

def system(t, x):
    return A @ x

t_span = (0, 10)
x0 = [1, 2]

sol = solve_ivp(system, t_span, x0, t_eval=np.linspace(0, 10, 100))

plt.plot(sol.t, sol.y[0], label='x1')
plt.plot(sol.t, sol.y[1], label='x2')
plt.xlabel('Time')
plt.ylabel('State')
plt.legend()
plt.title('Stability Analysis of Linear System')
plt.show()


#Markov Chain (Discrete-Time Approximation via Matrix Multiplication)

P = np.array([[0.9, 0.1],
              [0.5, 0.5]])

x = np.array([1, 0])  # Initial state
steps = 10

for i in range(steps):
    x = P @ x
    print(f"Step {i+1}: {x}")


#3. PCA (No ODE here, but eigen-decomposition)

data = np.array([[2.5, 2.4],
                 [0.5, 0.7],
                 [2.2, 2.9],
                 [1.9, 2.2],
                 [3.1, 3.0]])

# Center the data
data_centered = data - np.mean(data, axis=0)

cov_matrix = np.cov(data_centered, rowvar=False)

eigenvalues, eigenvectors = np.linalg.eigh(cov_matrix)

print("Eigenvalues:", eigenvalues)
print("Eigenvectors:\n", eigenvectors)


```

```python
#lec 8  Vector Spaces & Subspaces

'''
Definition of a Vector Space

Subspaces

Basis and Dimension

Row Space, Column Space, Null Space

Rank-Nullity Theorem


'''

import numpy as np
from scipy.linalg import null_space
import matplotlib.pyplot as plt

# Define matrix A
A = np.array([[1, 2],
              [3, 6]])

print("Matrix A:")
print(A)

# Row space basis (using np.linalg.matrix_rank and row echelon)
rank = np.linalg.matrix_rank(A)
print("\nRank of A:", rank)

# Column space basis (use np.linalg.svd or np.linalg.matrix_rank)
# For column space, find independent columns via SVD or QR
u, s, vh = np.linalg.svd(A)
print("\nSingular values:", s)

# Null space basis using scipy
ns = null_space(A)
print("\nNull space basis vectors (columns):")
print(ns)

# Dimension of null space
nullity = ns.shape[1]
print("\nNullity of A:", nullity)

# Check rank-nullity theorem
print(f"\nRank + Nullity = {rank} + {nullity} = {rank + nullity} (Number of columns: {A.shape[1]})")

# --- Visualization ---

# Visualizing the null space vectors in 2D
plt.figure(figsize=(6,6))
plt.axhline(0, color='gray')
plt.axvline(0, color='gray')

# Plot null space vector (only one vector here)
for i in range(ns.shape[1]):
    vec = ns[:, i]
    # Draw line through origin in direction of vec
    t = np.linspace(-5, 5, 100)
    plt.plot(t * vec[0], t * vec[1], label=f'Null space vector {i+1}')

plt.xlim(-5,5)
plt.ylim(-5,5)
plt.xlabel('x1')
plt.ylabel('x2')
plt.title('Null Space Visualization')
plt.legend()
plt.grid(True)
plt.show()
```

```python

#10. Matrix Factorizations & Decompositions

'''
#LU Decomposition

Factorizes a square matrix 
A into:A=LU
where
L = Lower triangular matrix (with 1s on diagonal),
U = Upper triangular matrix.
Used for solving linear systems 
Ax=b efficiently by forward and backward substitution.
Fast and widely used in numerical linear algebra.

#QR Decomposition

Factorizes A into:A=QR
whereQ is orthonormal,R is upper triangular.
Useful for least squares problems and eigenvalue algorithms.
More numerically stable than LU in some contexts.

#Cholesky Decomposition

Specialized decomposition for symmetric, positive-definite matrices:
A=LL^T
where L is lower triangular.
Faster and more efficient than LU for this special case.
Used in optimization, probabilistic models, and solving systems with covariance matrices.


#Singular Value Decomposition (SVD) – CRUCIAL in ML

Factorizes any m×n matrix A into:
A=UΣV^T where U and V are orthonormal matrices,Σ is diagonal with singular values.
Applications:
Dimensionality reduction (PCA),
Noise reduction,
Latent Semantic Analysis,
Recommender systems,
Image compression.
Handles any matrix, even non-square or rank-deficient.

#Eigen Decomposition

Decomposes a square matrix A as:
A=PDP^ −1
where D is diagonal (eigenvalues),P columns are eigenvectors.
Fundamental in:
Stability analysis,
Principal Component Analysis (PCA),
Dynamic systems,
Spectral clustering.

#Moore-Penrose Pseudo-Inverse:

Generalizes matrix inverse to non-square or singular matrices.
Denoted as A^+
Used to compute least squares solutions:
x=A^+b
Computed via SVD.
Important when A is not invertible or when solving under/over-determined systems.
'''
import numpy as np
from scipy.linalg import lu

A = np.array([[4, 3], [6, 3]])

# LU Decomposition
P, L, U = lu(A)
print("P:\n", P)
print("L:\n", L)
print("U:\n", U)

# QR Decomposition
Q, R = np.linalg.qr(A)
print("Q:\n", Q)
print("R:\n", R)

# Cholesky Decomposition (matrix must be symmetric positive-definite)
A_spd = np.array([[4, 2], [2, 3]])
L_chol = np.linalg.cholesky(A_spd)
print("Cholesky L:\n", L_chol)

# Singular Value Decomposition (SVD)
U_svd, S_svd, VT_svd = np.linalg.svd(A)
print("U (SVD):\n", U_svd)
print("Singular values:\n", S_svd)
print("V^T (SVD):\n", VT_svd)

# Eigen Decomposition
eigvals, eigvecs = np.linalg.eig(A_spd)
print("Eigenvalues:\n", eigvals)
print("Eigenvectors:\n", eigvecs)

# Moore-Penrose Pseudo-Inverse
A_pinv = np.linalg.pinv(A)
print("Pseudo-inverse:\n", A_pinv)

```
